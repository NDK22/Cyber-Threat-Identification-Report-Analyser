```python
import csv

# Technique IDs to extract
technique_ids_to_extract = [
    'T1003.001', 'T1210', 'T1570', 'T1140', 'T1218.011', 'T1059.003',
    'T1057', 'T1518.001', 'T1106', 'T1082', 'T1016', 'T1078', 'T1047',
    'T1027', 'T1056.001', 'T1083', 'T1053.005', 'T1070.004', 'T1105',
    'T1090', 'T1005', 'T1574.002', 'T1071.001', 'T1484.001',
    'T1204.002', 'T1055', 'T1562.001', 'T1033', 'T1566.001', 'T1219',
    'T1547.001', 'T1021.001', 'T1543.003', 'T1569.002', 'T1036.005',
    'T1112', 'T1041', 'T1110', 'T1190', 'T1564.001', 'T1113',
    'T1573.001', 'T1095', 'T1552.001', 'T1012', 'T1074.001',
    'T1548.002', 'T1068', 'T1072', 'T1557.001'
]

with open("/content/final_scrapped_techniques.csv") as file:
    lines = csv.reader(file)
    documents = []
    metadata = []
    ids = []

    for i, line in enumerate(lines):
        if i == 0:
            continue

        technique_id = line[0]

        if technique_id in technique_ids_to_extract:
            documents.append(line[2])
            metadata.append({"ID": technique_id})
            ids.append(str(len(ids) + 1))

# Now, 'documents', 'metadata', and 'ids' contain the extracted data for the specified technique IDs.
```

```python
import chromadb
from chromadb.utils import embedding_functions
client = chromadb.Client()
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-mpnet-base-v2")
collection = client.create_collection("technique_collection", embedding_function = sentence_transformer_ef)
```

```python
collection.add(
    documents = documents ,
    metadatas = metadata,
    ids = ids
)
```

```python
from langchain.llms import OpenAI
import json
import re
import pandas as pd

# Load your JSON file containing text and labels
json_file_path = '/content/single_label (1).json'  # Replace with the actual path to your JSON file

with open(json_file_path, 'r') as file:
    data = json.load(file)

# Create a DataFrame from the JSON data
df = pd.DataFrame(data)

# Randomly select a subset of query_texts from the DataFrame
num_query_texts = 100  # Replace with the desired number of query texts
selected_queries = df['text'].head(50).tolist()
#print("--------", selected_queries)

# Initialize variables for overall accuracy calculation
correct_predictions = 0
total_predictions = 0

technique_id_pattern = re.compile(r'Technique ID: (T\d+)')
# Query the vector database for each selected query text
for index, query_text in enumerate(selected_queries):
    llm = OpenAI(temperature=0)
    results = collection.query(query_texts=[question.format(query_text)], n_results=1, include=["documents", "metadatas"])
    #print(results)
    technique_str = ",".join(results["documents"][0])

    # Use the new OpenAI API interface
    response = llm.predict(
        text=f"System: {context.format(technique_str)}\nUser: {question}\n",  # Provide the 'text' argument here
        # model="text-davinci-003",  # or another suitable model
        # temperature=0,
        # max_tokens=100,
        # n=1,
    )

    # Assuming you have a 'label' column in your DataFrame
    actual_label = df['label'].iloc[index]


    match_1 = re.search(technique_id_pattern, response)

    if match_1:
        predicted_label = match_1.group(1).strip()
        print(predicted_label)
        print(actual_label)

        # Check if the predicted Technique ID matches the actual label
        if predicted_label == actual_label:
            correct_predictions += 1

        total_predictions += 1
# Calculate accuracy
accuracy = correct_predictions / total_predictions
print(f"Accuracy: {accuracy * 100:.2f}%")
```
